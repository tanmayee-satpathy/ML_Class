```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
n = len(X)

m = 0
c = 0
alpha = 0.01   
epochs = 1000

for i in range(epochs):
    y_pred = m*X + c
    dm = (-2/n) * sum(X * (y - y_pred))
    dc = (-2/n) * sum(y - y_pred)
    
    m = m - alpha * dm
    c = c - alpha * dc

print("Slope (m):", m)
print("Intercept (c):", c)

plt.scatter(X, y)
plt.plot(X, m*X + c, color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression using Gradient Descent")
plt.show()
```
